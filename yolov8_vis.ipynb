```python
# Táº£i má»™t mÃ´ hÃ¬nh YOLOv8 Ä‘Ã£ Ä‘Æ°á»£c Ä‘Ã o táº¡o trÆ°á»›c (vÃ­ dá»¥: yolov8n.pt cho phiÃªn báº£n nano)
# Báº¡n cÃ³ thá»ƒ thay Ä‘á»•i 'yolov8n.pt' thÃ nh 'yolov8s.pt', 'yolov8m.pt', 'yolov8l.pt', 'yolov8x.pt' hoáº·c má»™t tá»‡p trá»ng sá»‘ tÃ¹y chá»‰nh náº¿u báº¡n cÃ³.
from ultralytics import YOLO

# Táº£i mÃ´ hÃ¬nh YOLOv8
model = YOLO('yolov8n.pt')

print("MÃ´ hÃ¬nh YOLOv8 Ä‘Ã£ Ä‘Æ°á»£c táº£i thÃ nh cÃ´ng!")
```

### Huáº¥n luyá»‡n mÃ´ hÃ¬nh YOLOv8

Báº¡n cÃ³ thá»ƒ thay Ä‘á»•i cÃ¡c tham sá»‘ nhÆ° `epochs` (sá»‘ lÆ°á»£ng epoch huáº¥n luyá»‡n), `imgsz` (kÃ­ch thÆ°á»›c hÃ¬nh áº£nh Ä‘áº§u vÃ o), `batch` (kÃ­ch thÆ°á»›c lÃ´) vÃ  `name` (tÃªn cho láº§n cháº¡y huáº¥n luyá»‡n nÃ y).

```python

results = model.train(data='/content/drive/MyDrive/data/visdrone_small.yaml', epochs=50, imgsz=640, name='yolov8n_custom_dataset')

print("Huáº¥n luyá»‡n Ä‘Ã£ hoÃ n táº¥t!")
```

```python
path_to_best_weights = '/content/runs/detect/yolov8n_custom_dataset2/weights/best.pt'

# Táº£i mÃ´ hÃ¬nh Ä‘Ã£ huáº¥n luyá»‡n
model = YOLO(path_to_best_weights)

results = model.predict(source='/content/drive/MyDrive/VisDrone2019-DET-test-dev/images', save=True, conf=0.25, iou=0.7)

print("Kiá»ƒm tra Ä‘Ã£ hoÃ n táº¥t! Káº¿t quáº£ Ä‘Æ°á»£c lÆ°u trong thÆ° má»¥c 'runs/detect/predict'.")
```

```python
# Táº£i mÃ´ hÃ¬nh tá»‘t nháº¥t Ä‘Ã£ huáº¥n luyá»‡n cá»§a báº¡n
# Äáº£m báº£o ráº±ng `path_to_best_weights` trá» Ä‘áº¿n tá»‡p trá»ng sá»‘ `best.pt` cá»§a báº¡n.
# VÃ­ dá»¥: path_to_best_weights = '/content/runs/detect/yolov8n_custom_dataset2/weights/best.pt'
model = YOLO(path_to_best_weights) # Sá»­ dá»¥ng mÃ´ hÃ¬nh Ä‘Ã£ táº£i tá»« bÆ°á»›c predict hoáº·c táº£i láº¡i náº¿u báº¡n muá»‘n

# Thá»±c hiá»‡n Ä‘Ã¡nh giÃ¡
# Thay tháº¿ '/content/drive/MyDrive/data/visdrone_small.yaml' báº±ng Ä‘Æ°á»ng dáº«n thá»±c táº¿ Ä‘áº¿n tá»‡p data.yaml cá»§a báº¡n
metrics = model.val(data='/content/drive/MyDrive/data/visdrone_small.yaml')

# In ra cÃ¡c chá»‰ sá»‘ chÃ­nh
print("mAP50-95:", metrics.box.map)
print("mAP50:", metrics.box.map50)
print("mAP75:", metrics.box.map75)
print("Äá»™ chÃ­nh xÃ¡c (Precision):", metrics.box.mp)
print("Äá»™ thu há»“i (Recall):", metrics.box.mr)

print("ÄÃ¡nh giÃ¡ Ä‘Ã£ hoÃ n táº¥t! CÃ¡c káº¿t quáº£ chi tiáº¿t hÆ¡n (nhÆ° biá»ƒu Ä‘á»“, ma tráº­n nháº§m láº«n) Ä‘Æ°á»£c lÆ°u trong thÆ° má»¥c 'runs/detect/'.")
```
